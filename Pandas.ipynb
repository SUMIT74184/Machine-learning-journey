# Working with the time

```python
import pandas as pd
from datetime import datetime, timedelta

names = [
    "John", "Emma", "Oliver", "Sophia", "Liam", "Ava", "Noah", "Mia",
    "William", "Isabella", "James", "Charlotte", "Benjamin", "Amelia",
    "Lucas", "Harper", "Henry", "Evelyn", "Alexander", "Abigail",
    "Michael", "Ella", "Daniel", "Scarlett", "Matthew", "Grace",
    "Joseph", "Chloe", "David"
]

start_date = datetime(2025, 1, 1)

rows = []
for i in range(1, 30):
    rows.append({
        "index": i,
        "date": (start_date + timedelta(days=i-1)).strftime("%Y-%m-%d"),
        "customername": names[i-1],
        "lemonades_count": i * 2
    })

df = pd.DataFrame(rows)
```

```python
df
```

```python
df.drop(columns='index',inplace=True)
```

```python
df
```

```python
df['date'] = pd.to_datetime(df['date'])# coeecting the correct format
```

```python
df
```

```python
df.set_index('date', inplace=True)# Setting the index to date col
```

```python
df
```

```python

pd.to_datetime('2025-04-29')
```

```python
#Resampling
# df.resample('W').sum()
# df.resample('ME').sum()
df.resample('YE').sum()
```

```python
#Shifting/Lagging
df['Previous Lemonades'] = df['lemonades_count'].shift(1)
```

```python
df
```

```python
#Rolling Calculation
df['Rolling Average'] = df['lemonades_count'].rolling(window=2).mean()
```

```python
df
```
