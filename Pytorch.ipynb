# Matrix Multiplication

Two main ways of performing multiplication in neural networks

1.Element-wise multiplication

2.Matrix multipication (dot product)

There are two main rules that performing matrix multiplication needs to satisfy:

  1.The **inner dimensions** must match:

    * `(3,2) @ (3,2)` won't work
    * `(2,3) @ (3,2)` will work
    * `(3,2) @ (2,3)` will work
  
  2. The resulting matrix has the shape of the **outer dimensions**:
    
    * `(2,3) @ (3,2)` -> (2,2)
    * `(3,2) @ (2,3)` -> (3,3)


```python
# torch.matmul(torch.rand(3,2),torch.rand(3,2))  mat1 and mat2 shapes cannot be multiplied (3x2 and 3x2)
torch.matmul(torch.rand(3,2),torch.rand(2,3))
```

```python
torch.matmul(torch.rand(2,3),torch.rand(3,2))
```

```python
#randomly defining the matrix
torch.matmul(torch.rand(3,1),torch.rand(1,3))
```

# Matrix multiplication dealing with tensor shape Errors

```python
# shape for matrix multiplication
tensor_A=torch.tensor([[1,2],
                      [3,4],
                      [5,6]])

tensor_B=torch.tensor([[7,10],
                       [8,11],
                       [9,11]])
torch.mm(tensor_A,tensor_B)

```

```python
tensor_A.shape,tensor_B.shape
```

To fix our tensor shape issues, we can manipulate the shape of one of our tensors using a transpose

A **transpose** switches the axes or dimensions of a given tensors

```python
tensor_B.T
```

```python
tensor_B.T.shape
```

```python
#original tensor_B
tensor_B
```

```python
tensor_B.shape
```

```python
torch.matmul(tensor_B.T,tensor_B)
```

```python
torch.matmul(tensor_B,tensor_B.T)
```

```python
#The matrix multiplication operation works when tensor_B is transposed
print(f"Original shapes: tensor_A= {tensor_A.shape}, tensor_B = {tensor_B.shape}")
print(f"New Shapes: tensor_A = {tensor_A.shape} (same shape as above), tensor_B.T = {tensor_B.T.shape}")
print(f"Multiplying:{tensor_A.shape} @ {tensor_B.T.shape} <-inner dimensions must match")
print("Output:\n")
output = torch.matmul(tensor_A,tensor_B.T)
print(output)
print(f"\noutput shape: {output.shape}")
```

## Reshaping,stacking,squeezing and unsequeezing tensors
* Reshaping - reshapes an input tensor to a defined shape
* View- Return a view of an input tensor of certain shape but keep the same memory as the original tensor
* Stacking - combine multiple tensors on top of each other (vstack) or side by side (hstack)
* Squeeze - removes all `1` dimensions from a tensor
* Unsqueeze - add a `1` dimensions to a target tensor
* Permute - Return a view of the input with dimensions permuted(Swapped) in certain way

```python
import torch
x=torch.arange(1.,11.)
```

```python
x,x.shape
```

```python
# Add an extra dimension
x_reshaped=x.reshape(10,1)
x_reshaped
```

```python
x_reshaped=x.reshape(1,10)
x_reshaped
```

```python
# Change the view
z=x.view(1,10)
z,z.shape
```

```python
# Changing z changes x (because a view of a tensor shares the same memory as the original)
z[:,0] = 5
z,x
```

```python
# stacks tensors on top of each other
x_stacked = torch.stack([x,x,x,x],dim=1)
x_stacked
```

```python
# torch.squeeze()
print(f"Previous tensor{x_reshaped}")
print(f"Previos shape{x_reshaped.shape}")

#remove extra dimensions from x_reshaped

x_squeezed=x_reshaped.squeeze()
print(f"\nNew tensor:{x_squeezed}")
print(f"\nNew Shape:{x_squeezed.shape}")


# x_reshaped
```

```python
# torch.unsqueeze() - adds a single dimension to a target tensor at a specific dim (dimension)
print(f"Previous target:{x_squeezed}")
print(f"previos shape:{x_squeezed.shape}")

#Add an extra dimension with unsqeeze
x_unsqueezed=x_squeezed.unsqueeze(dim=0) # try dim=1
print(f"\nNew tensor:{x_unsqueezed}")
print(f"New Shape: {x_unsqueezed.shape}")
```

```python
#torch.permute()- reshapes the dimensions of a target tensor is specified order
x_original = torch.rand(size=(224,224,3)) # [height, width, color_channels]

#permute the original tensor to rearange the axis (or dim) order
x_permuted= x_original.permute(2,0,1) #shifts axis 0->1 ,1->2 2->0

print(f"previos shape:{x_original.shape}")
print(f"New Shape:{x_permuted.shape}") #color_channels, height, width
```

# Pytorch tensors & NumPy

NumPy is a popular scientific python numerical computing library

And because of this,pytorch has functionality to interact with it.

* Data in NumPy,want in Pytorch tensor -> `torch.from_numpy(ndarray)`
* Pytorch tensor ->NumPy -> `torch.Tensor.numpy()`

```python
import torch
import numpy as np

array=np.arange(1.0,8.0)

tensor=torch.from_numpy(array) # warning: when converting from numpy-> pytorch,pytorch reflects numpy's default datatype of float64 unless specified
array,tensor
```

## Reproducibilty (trying to take random out of random)

In short how neural networks learns:

`start with random numbers -> tensor operations -> update random numbers to try and make them of data -> again -> again -> again....`

To reduce the randomness in neural networks and pytorch comes the concept of **random seed**

Essentially what the random seed does is "flavour" the randomness

```python
torch.rand(2,2)
```

```python
import torch
```

```python
#create two random tensors
random_tensor_A=torch.rand(3,4)
random_tensor_B= torch.rand(3,4)
```

```python
1==1
```

```python
print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
```

```python
# Let's make some random but reproducible tensors

#Set the random seed
RANDOM_SEED=43
torch.manual_seed(RANDOM_SEED)

random_tensor_C=torch.rand(3,4)
random_tensor_D=torch.rand(3,4)

print(random_tensor_C)
print(random_tensor_D)

print(random_tensor_C==random_tensor_D)
```
